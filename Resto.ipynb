# Analyse des données :

```python
a =(df['item_name']=='Canned Soda').sum()
print('Boissons gazeuse :', a)
b= (df['item_name']=='Bottled Water').sum()
print('Eau en bouteille :', b)
```

Les clients privilégient les boissons saines. Ils consomment de l'eau plus que les boissons gazeuses.

```python
#les trois boissons gazeuses les plus commandées :
df3 = df['choice_description'].value_counts().sort_values(ascending=False)[0:3]
top_3 = df3.head(3)
top_3
```

```python
ax = top_3.plot(kind='bar')
ax.set_xticklabels(top_3.index, rotation=45)
```

Les boissons gazeuses les plus consommées par les clients sont le Diet Coke, le Coke et le Sprite.

La popularité du Diet Coke parmi ces boissons, reflète toujours une tendance vers une consommation plus saine.

```python
#Les 6 plats les plus populaires
df1 = df['item_name'].value_counts().sort_values(ascending=False)[:4]
df1
```

```python
#Graphique des 3 plats les plus populaires
ax = df1.plot(kind='bar')
ax.set_xticklabels(df1.index, rotation=45)
```

Les trois plats les plus populaires sont le Chicken Bowl, le Chicken Burrito et les Chips et Guacamole.

Les clients préfèrent les plats a base de poulets et les accompagnements tels que les chips et le guacamole.


```python
df['item_price'] =  pd.to_numeric(df['item_price'].replace('\$', ''))
```

```python
sns.scatterplot(data=df,x='item_price',y='quantity' )
```

On peut observer une corrélation entre le prix des items et la quantité commandée.

Les items les moins chers sont plus commandés que les items les plus chers.

```python
g =sns.scatterplot(data=df,x='item_name',y='item_price' )
xtick_labels = df['item_name'][::5]
g.set_xticklabels(xtick_labels, rotation=90)
plt.show()

```

Il existe une grande variabilité dans les prix des plats.

Un plat peut avoir des prix différents parce que les clients prsonnalisent leurs commandes.
